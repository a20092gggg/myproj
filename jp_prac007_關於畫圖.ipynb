# 關於array的操作
* xy = [[x,y] for x in range() for y in range()]
* 單位矩陣語法:np.eye()
* 快篩array的語法:array=array[條件],這樣會直接篩出來。就不用用迴圈判別式了。

# plot的方法
* plt.plot()  ，後面可以直接放array，或是類似matlab的 plot(x,y)
* 設定座標範圍語法:np.linspace(A,B,C) 從A到B之間取C個點
* 畫圖可以直接疊上去，不用在那邊hold on
* 利用array快篩的方法，也可以用在畫圖上。
* 語法：plt.plot(x[條件],y[條件],線的種類),前後條件要一致。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
xy = [[x,y] for x in range(10) for y in range(5)]
```

```python
xy = np.array(xy)
```

```python
np.eye(10)
```

```python
y = np.random.randn(50)
plt.plot(y)
```

```python
plt.plot(xy)
```

```python
x = np.linspace(0,np.pi*8,700)
plt.plot(x,np.sin(x))
```

```python
y = np.sinc(x)
plt.plot(x,y)
plt.plot(x[y>0],y[y>0],'.')
```
