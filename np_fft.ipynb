```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(0,360)
x = np.arange(-3000,3000)
y1 = np.exp(x1/100-2*np.pi)
y2 = np.cos(x1 * np.pi / 180.0)
```

```python
plt.subplot(2,2,1)
plt.plot(np.fft.fft(y1,1000))

plt.subplot(2,2,2)
plt.plot(np.fft.fft(y2,1000))
yyy = np.fft.ifft(np.fft.fft(y1,800) * np.fft.fft(y2,800) )

plt.subplot(2,2,3)
plt.plot(np.convolve(y1,y2))

plt.subplot(2,2,4)
plt.plot(yyy)
```

```python
def flip180(arr):
    new_arr = arr.reshape(arr.size)
    new_arr = new_arr[::-1]
    new_arr = new_arr.reshape(arr.shape)
    return new_arr
```
