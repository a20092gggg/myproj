# array的功用
* list是不能乘的，但陣列可以(想想matlab)
* 所以要做的事情就是，把list 轉乘 array
* 語法: np.array(_)
* 內積語法: np.dot(_,_)
* array.shape = (_,_) 可以看array的形狀，後面可以改變陣列的排列
* 改變陣列也可以用 變數=array.reshape(_,_)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
p=[1096.95,596.95,896.95]
c=30.1
result=[]
for k in p:
    twd=k*c
    result.append(twd)
```

```python
print(result)
```

```python
grades = np.array([93,93,92,90,96,88,95])
weights = np.array([2,2,2,2,2,2,2])
```

```python
np.dot(grades,weights)/ np.sum(weights)  # 權重的寫法。
```

```python
grade = np.array([[93,92,95],[95,98,89],[55,12,58],[98,78,64]])
w = np.array([0.4,0.3,0.3])
np.dot(grade,w)    #多人平均加權的算法，把array寫成多個list
```
